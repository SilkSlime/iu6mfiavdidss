```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/flight_details.csv")
data = df.pivot(index="Months", columns="Years", values="Passengers")

# Используем желтую палитру
sns.heatmap(data, cmap="YlGn", annot=True, fmt='d')

plt.title("Flight Passengers from 2001 to 2012")
plt.show()
```

```python
df = pd.read_csv("data/movie_scores.csv")

movie_scores = pd.DataFrame({
    "Movie Title": list(df["MovieTitle"]) * 2,
    "Score": list(df["AudienceScore"]) + list(df["Tomatometer"]),
    "Type": ["Audience Score"] * len(df["AudienceScore"]) + ["Tomatometer"] * len(df["Tomatometer"])
})

# sns.set()
# Устанавливаем розовую палитру
sns.set_palette("pink")

ax = sns.barplot(data=movie_scores, x="Movie Title", y="Score", hue="Type")
plt.xticks(rotation=10)
plt.title("Movies Scores comparison")
plt.xlabel("Movies")
plt.ylabel("Scores")
plt.show()
```

```python
df = pd.read_csv("data/water_usage.csv")

import squarify

labels = df['Usage'] + ' (' + df['Percentage'].astype('str') + '%)'
squarify.plot(sizes=df['Percentage'], label=labels, color=sns.light_palette('pink', df.shape[0]))
plt.axis('off')
plt.title('Water usage')
plt.show()
```
