```python
import matplotlib.pyplot as plt
import random

# Генерируем случайное количество ступенек
num_steps = random.randint(5, 15)
step_length = 1  # Длина первой ступени

# Инициализируем список для хранения координат точек
x = [0]  # начинаем с нулевой координаты X
y = [0]  # начинаем с нулевой координаты Y

# Создаем координаты для ступенек
for i in range(num_steps):
    x.extend([i, i + 1])
    y.extend([i * step_length, (i + 1) * step_length])
    step_length *= 2  # Увеличиваем длину для следующей ступени

# Создаем Scatter Plot
plt.scatter(x, y, marker='j', s=50, color='orange')  # Используем квадратные маркеры размером 100

# Создаем пунктирную линию
for i in range(1, len(x), 2):
    plt.plot(x[i-1:i+1], y[i-1:i+1], linestyle='--', color='blue')

    
# Настройка графика
plt.xlabel('Номер ступеньки')
plt.ylabel('Высота ступеньки')
plt.title('Лестница из случайного количества ступенек')

# Отображаем график
plt.show()
```

```python
import matplotlib.pyplot as plt

# Имитированные данные
continents = ['Азия', 'Европа', 'Африка', 'Америка', 'Австралия']
countries = [45, 28, 33, 29, 7]  # Количество стран в каждом континенте

# Настройка цветов и расстояния между частями
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.05, 0.05, 0.05, 0.05, 0.05)  # Добавляем расстояние между секторами

# Создаем круговую диаграмму
plt.figure(figsize=(8, 8))
plt.pie(countries, labels=continents, autopct='%1.1f%%', startangle=200, colors=colors, explode=explode,
        labeldistance=1.05)  # Используем labeldistance для настройки расстояния между подписями и секторами
plt.axis('equal')  # Делает круговую диаграмму круглой

plt.title('Иерархия континентов и стран')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Имитированные данные
np.random.seed(0)
heights = np.random.uniform(150, 190, 30)
weights = np.random.uniform(50, 100, 30)
ages = np.random.uniform(18, 22, 30)

# Создаем Bubble plot
plt.figure(figsize=(10, 8))
plt.scatter(heights, weights, c=ages, cmap='viridis', s=ages*10, alpha=0.7)
plt.xlabel('Рост (см)')
plt.ylabel('Вес (кг)')
plt.title('Зависимость между ростом, весом и возрастом')
plt.colorbar(label='Возраст')
plt.grid(True)
plt.show()
```

```python
# Создаем Box plot
plt.figure(figsize=(10, 8))
data = [heights, weights, ages]
labels = ['Рост', 'Вес', 'Возраст']
plt.boxplot(data, labels=labels)
plt.ylabel('Значение')
plt.title('Статистика по росту, весу и возрасту')
plt.grid(axis='y')
plt.show()
```
