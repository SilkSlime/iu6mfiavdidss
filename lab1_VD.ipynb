```python
import pandas as pd

df = pd.read_csv('data/homicide-data.csv', encoding = 'ISO-8859-1')
df['victim_age'] = pd.to_numeric(df['victim_age'], errors='coerce')

print("Информация о первых 5 пассажирах:")
df.head(5)
```

```python
print("Информация о всех пассажирах, кроме последних двух:")
df[:-2]
```

```python
print("Информация о пассажирах с 3 по 15:")
df[2:15]
```

```python
# Средний возраст, максимальный и минимальный

average_age = df['victim_age'].mean()
max_age = df['victim_age'].max()
min_age = df['victim_age'].min()
print(f'Средний возраст: {average_age:.1f} лет')
print(f'Максимальный возраст: {max_age:.1f} лет')
print(f'Минимальный возраст: {min_age:.1f} лет')
```

```python
# Количество женщин и мужчин
female_count = df[df['victim_sex'] == 'Female'].shape[0]
male_count = df[df['victim_sex'] == 'Male'].shape[0]
print(f'Количество женщин: {female_count}')
print(f'Количество мужчин: {male_count}')
```

```python
# Количество женщин старше 18 лет
female_over_18_count = df[(df['victim_sex'] == 'Female') & (df['victim_age'] > 18)].shape[0]
print(f'Количество женщин старше 18 лет: {female_over_18_count}')
```

```python
# Количество мужчин, которые были в первом классе (disposition == Closed by arrest)
first_class_male_count = df[(df['disposition'] == 'Closed by arrest') & (df['victim_sex'] == 'Male')].shape[0]
print(f'Количество мужчин в первом классе: {first_class_male_count}')
```

```python
sorted_ascending = df.sort_values(by='victim_age', ascending=True)
sorted_descending = df.sort_values(by='victim_age', ascending=False)
print("Сортировка по возрасту по возрастанию:")
sorted_ascending
```

```python
print("Сортировка по возрасту по убыванию:")
sorted_descending
```
