```python
import matplotlib.pyplot as plt

# Имитированные данные о посещаемости (проценты)
week1_attendance = [95, 80, 70, 100, 60, 90, 75, 85, 88, 92]
week2_attendance = [90, 85, 68, 98, 70, 92, 80, 84, 90, 94]

# Список студентов (номера)
students = list(range(1, len(week1_attendance) + 1))


# Создаем гистограмму посещаемости за 2 недели
plt.figure(figsize=(10, 6))
plt.hist([week1_attendance, week2_attendance], bins=10, label=['Неделя 1', 'Неделя 2'], alpha=0.7)
plt.xlabel('Процент посещаемости')
plt.ylabel('Количество студентов')
plt.title('Гистограмма посещаемости за 2 недели')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Создаем Bar Chart посещаемости за 2 недели
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.4
bar1 = np.arange(len(week1_attendance))
bar2 = [x + bar_width for x in bar1]
ax.bar(bar1, week1_attendance, width=bar_width, label='Неделя 1', alpha=0.7)
ax.bar(bar2, week2_attendance, width=bar_width, label='Неделя 2', alpha=0.7)

# Настройка графика
ax.set_xlabel('Студенты')
ax.set_ylabel('Процент посещаемости')
ax.set_title('Bar Chart посещаемости за 2 недели')
ax.set_xticks([x + bar_width/2 for x in bar1])
ax.set_xticklabels([str(i + 1) for i in range(len(week1_attendance))])
ax.legend()
ax.grid(True)

# Отображаем оба графика
plt.show()
```

```python
# Гистограмма (histogram):

# Преимущества: Гистограмма хорошо подходит для визуализации распределения данных. Она позволяет увидеть, как данные распределены по различным диапазонам или категориям (в данном случае, по процентам посещаемости). Гистограмма показывает частоту встречаемости каждого значения или диапазона значений.
# Почему удобно: Гистограмма подходит, если вы хотите оценить, насколько равномерно распределена посещаемость студентов и увидеть, сколько студентов имеют похожий процент посещаемости. Это позволяет легко сравнивать общее распределение данных.
    
# Bar Chart:

# Преимущества: Bar Chart хорошо подходит для сравнения значений между разными категориями (в данном случае, между студентами и двумя неделями). Он позволяет увидеть, какие студенты имеют более высокую посещаемость в разные недели.
# Почему удобно: Bar Chart подходит, если вы хотите сравнить посещаемость каждого студента за две недели и выявить изменения. Это позволяет увидеть, какие студенты более стабильны в посещении и какие менее стабильны.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Имитированные данные для 3 уровней лестницы
level1 = [0, 2, 4, 6, 8]
level2 = [0, 1, 3, 5, 7]
level3 = [0, 1, 2, 4, 6]

# Создаем Stacked Area Chart
plt.figure(figsize=(10, 6))

plt.fill_between(range(5), np.array(level1) + np.array(level2) + np.array(level3), label='Уровень 3', alpha=1)
plt.fill_between(range(5), np.array(level1) + np.array(level2), label='Уровень 2', alpha=1)
plt.fill_between(range(5), level1, label='Уровень 1', alpha=1)

# Настройка графика
plt.xlabel('Шаг')
plt.ylabel('Высота')
plt.title('Stacked Area Chart лестницы с 3 уровнями')
plt.legend(loc='upper left')

plt.show()
```
